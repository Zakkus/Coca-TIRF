```python
def grayscale(img):
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    for i in range(0,h-1):
        for j in range(0,w-1):
            (r,g,b) = arr[i,j]
            gray = 0.299 * r + 0.587 * g + 0.114 * b
            arr[i,j] = (gray,gray,gray)
    return arr
```

```python
def RGBtoHSV(rgb):
    (r,g,b) = rgb
    t = 0
    maxi = max(r,g,b)
    mini = min(r,g,b)
    if maxi == mini:
        t = 0
    elif maxi == r:
        t = (60 * (g-b)/(maxi - mini) + 360) % 360
    elif maxi == g:
        t = 60 * (b-r)/(maxi - mini) + 120
    elif maxi == b:
        t = 60 * (r-g)/(maxi - mini) + 240
    v = maxi
    s = 0
    if maxi > 0:
        s = 1 - mini/maxi
    hsv = (t,s,v)
    return hsv

def HSVtoRGB(hsv):
    (h,s,v) = hsv
    t = (h/60) % 6
    f = h/60 - t
    l = v * (1 - s)
    m = v * (1 - f * s)
    n = v * (1- (1 - f) * s)
    if t == 0:
        rgb = (v,n,l)
    elif t ==1:
        rgb = (m,v,l)
    elif t == 2:
        rgb = (l,v,n)
    elif t == 3:
        rgb = (l,m,v)
    elif t == 4:
        rgb = (n,l,v)
    else:
        rgb = (v,l,m)
    return rgb
```

```python
def RGBtoHSL(rgb):
    (r,g,b) = rgb
    t = 0
    maxi = max(r,g,b)
    mini = min(r,g,b)
    if maxi == mini:
        t = 0
    elif maxi == r:
        t = (60 * (g-b)/(maxi - mini) + 360) % 360
    elif maxi == g:
        t = 60 * (b-r)/(maxi - mini) + 120
    elif maxi == b:
        t = 60 * (r-g)/(maxi - mini) + 240
    l = (maxi + mini)/2
    s = 0
    if maxi != mini:
        s = (maxi-mini)/(1 - fabs(2 * l - 1))
    return (t,s,l)

def HSLtoRGB(hsl):
    (h,s,l) = hsl
    t = (h/60) % 6
    f = h/60 - t
    c = v * (1 - s)
    m = v * (1 - f * s)
    n = v * (1- (1 - f) * s)
    if t == 0:
        rgb = (v,n,l)
    elif t ==1:
        rgb = (m,v,l)
    elif t == 2:
        rgb = (l,v,n)
    elif t == 3:
        rgb = (l,m,v)
    elif t == 4:
        rgb = (n,l,v)
    else:
        rgb = (v,l,m)
    return rgb
```

```python
def red_filter(img):
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    for i in range(0,h-1):
        for j in range(0,w-1):
            (r,g,b) = arr[i,j]
            if g <=100 and b <= 100 and r > 100:
                arr[i,j] = (255,0,0)
            elif g >= 200 and b >= 200 and r >= 200:
                arr[i,j] = (255,255,255)
            else:
                arr[i,j] = (0,0,0)
    return arr

def full_red(img):
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    rbar = 0
    for i in range(0,w-1):
        for j in range(0,h-1):
            (r,g,b) = arr[i,j]
            rbar = rbar + r
            arr[i,j] = (r,0,0)
    rbar = rbar / (w*h)
    return (rbar,arr)

def mid_red(img, rbar):
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    for i in range(0,w-1):
        for j in range(0,h-1):
            (r,g,b) = arr[i,j]
            if r <= rbar:
                arr[i,j] = (0,g,b)
    return arr

def getall_mid(img):
    rbar = 0
    gbar = 0
    bbar = 0
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    for i in range(0,h-1):
        for j in range(0, w-1):
            (r,g,b) = arr[i,j]
            rbar = rbar + r
            gbar = gbar + g
            bbar = bbar + b
    rbar = rbar / (w*h)
    gbar = gbar / (w*h)
    bbar = bbar / (w*h)
    return (rbar, gbar, bbar)
```

```python
def mid(img,rgb):
    (rbar, gbar, bbar) = rgb
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    for i in range(0,h-1):
        for j in range(0,w-1):
            (r,g,b) = arr[i,j]
            if r <= rbar:
                arr[i,j] = (0,0,0)
            elif r > rbar and g > gbar or b > bbar:
                arr[i,j] = (255,255,255)
            else:
                arr[i,j] = (255,0,0)
    return arr

def boostCol(img):
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    for i in range(0,h-1):
        for j in range(0,w-1):
            rgb = arr[i,j]
            hsv = RGBtoHSV(rgb)
            hsv = boostV(hsv)
            rgb = HSVtoRGB(hsv)
            (r,g,b) = rgb
            arr[i,j] = (r,g,b)
    return arr
```

```python
def boostV(hsv):
    (h,s,v) = hsv
    if v > 50:
        v = 255
    hsv = (h,s,v)
    return hsv
```

```python
def chooseMax(img):
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    for i in range(0,h-1):
        for j in range(0,w-1):
            (r,g,b) = arr[i,j]
            if max(r,g,b) == r and max(r,g,b) == g and max(r,g,b) == b:
                arr[i,j] = (255,255,255)
            elif max(r,g,b) == r:
                arr[i,j] = (255,0,0)
            elif max(r,g,b) == g:
                arr[i,j] = (0,255,0)
            elif max(r,g,b) == b:
                arr[i,j] =(0,0,255)
    return arr
```

```python
#image en niveau de gris obligatoire
def Sobel(img):
    (w,h) = img.size
    arr = np.array(img, dtype=np.uint8)
    tmp = np.array(img, dtype=np.uint8)
    x_op = np.matrix('-1 0 1; -2 0 2; -1 0 1')
    y_op = np.matrix('1 2 1; 0 0 0; -1 -2 -1')
    for i in range(0,h-1):
        for j in range(0,w-1):
            sx = 0
            sy = 0
            for x in range(-1,1):
                for y in range(-1,1):
                    (r,g,b) = tmp[i + x, j + y]
                    sx = sx + r * x_op[x + 1,y + 1]
                    sy = sy + r * y_op[x + 1,y + 1]
            gray = 255 -(abs(sx) + abs(sy))
            arr[i,j] = (gray, gray, gray)
    return arr
```
